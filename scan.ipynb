```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('page22.tif')
#img = cv2.cvtColor(img, cv2.cv.CV_BGR2RGB)
```

```python
board = img[1045:2245, 1858:3058]
```

```python
plt.imshow(board)
```

```python
def get_cell(x, y):
    x_cell_size = 149
    y_cell_size = 150
    x *= x_cell_size
    y *= y_cell_size
    field = board[y:y+y_cell_size, x:x+x_cell_size]
    plt.imshow(field)
    return field
```

```python
f = get_cell(0,0)
```

```python
f = get_cell(7,7)
```

```python
dark = get_cell(3,4)
```

```python
field3 = board[0:300, 0:300]
plt.imshow(field3)
```

```python
_, w, h = dark.shape[::-1]
board_copy = board.copy()

res = cv2.matchTemplate(board, dark, cv2.TM_CCOEFF_NORMED)
threshold = 0.7
loc = np.where(res >= threshold)
for pt in zip(*loc[::-1]):
    cv2.rectangle(board_copy, pt, (pt[0] + w, pt[1] + h), (0,0,255), 2)

plt.imshow(board_copy)
```
